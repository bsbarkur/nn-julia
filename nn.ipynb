<h2>Custom helper dictionary that gets us the counts of all classes</h2>

```julia
classDict = Dict{Any,Any}()
for element in train_pd[95]
    if haskey(classDict, element) == true
        classDict[element] += 1
    else
        classDict[element] = 1
    end
end

classDict
```
